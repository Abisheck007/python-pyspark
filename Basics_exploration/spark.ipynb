```python
#rename
df1.withColumnRenamed("age","age of person").show()
```

```python
#to drop null values
df1.na.drop().show()
```

```python
#fill the missing values
df1.na.fill("no age given","age").show()#we are passing value to fill null and the column name..we can also pass list of columns
```

```python
#filling using mean,median,mode
from pyspark.ml.feature import Imputer

# Cast columns to double
df1_casted = df1.withColumn("age", df1["age"].cast("double")).withColumn("bmi", df1["bmi"].cast("double"))

imputer = Imputer(
    inputCols=["age", "bmi"],
    outputCols=["{}_imputed".format(c) for c in ["age", "bmi"]]
).setStrategy("mean")

imputer.fit(df1_casted).transform(df1_casted).show()
```

```python
#filter operation
df1.filter("age>25").select(["age","bmi"]).show()
```

```python
df1.filter((df1["age"].cast("double") > 26) & (df1["bmi"].cast("double") > 20)).select(["age","bmi"]).show()
```

```python
df1.filter(~(df1["age"].cast("double") > 26) & (df1["bmi"].cast("double") > 20)).select(["age","bmi"]).show()
```

```python
from pyspark.ml.feature import VectorAssembler
featureassembler = VectorAssembler(
	inputCols=["age", "bmi"],
	outputCol="Independent Features",
	handleInvalid="skip"  # Skip rows with nulls
)
output = featureassembler.transform(df1_casted)
output.select("Independent Features").show()
```
